```python
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import re

point = (33, 135)

# ジオコーダーの初期化
geolocator = Nominatim(user_agent="geoapi")

# 住所
full_address = "東京都板橋区中台3丁目22-18"
# 数字の前までを抽出（数字が始まる位置を探してスライス）
match = re.search(r'\d', full_address)
if match:
    address = full_address[:match.start()]
else:
    address = full_address

# 緯度経度の取得
location = geolocator.geocode(address)

if location:
    print("住所:", address)
    print("緯度:", location.latitude)
    print("経度:", location.longitude)
else:
    print("住所が見つかりませんでした。")

coord = (location.latitude, location.longitude)

# 距離を計算（単位：km）
distance_km = geodesic(point, coord).km
print(distance_km)
```
